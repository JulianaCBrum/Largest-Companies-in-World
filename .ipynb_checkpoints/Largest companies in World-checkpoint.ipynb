```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Largest companies in world.csv')
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
def convert_revenue_to_float(revenue):
    if pd.isna(revenue):  # Check if revenue value is NaN
        return revenue
    elif 'B' in revenue:  # If revenue is in billions
        return float(revenue.replace('B', '').replace(',', '')) * 1e9  # Convert to float (billions)
    elif 'M' in revenue:  # If revenue is in millions
        return float(revenue.replace('M', '').replace(',', '')) * 1e6  # Convert to float (millions)
    else:  # If revenue is already a number
        return float(revenue.replace(',', ''))  # Convert to float

# Apply the function to the 'revenue' column
df['revenue'] = df['revenue'].apply(convert_revenue_to_float)
```

```python
df.info()
```

```python
def convert_to_common_unit(value):
    if isinstance(value, float):  # Check if value is already a float
        return value
    elif 'B' in value:  # Check if value is in billions
        return float(value.replace('B', '').replace(',', '')) * 1000000000  # Convert to billions
    elif 'M' in value:  # Check if value is in millions
        return float(value.replace('M', '').replace(',', '')) * 1000000  # Convert to millions
    else:
        return float(value.replace(',', ''))  # If no unit is specified, convert value to float

# Apply the function to revenue, profits, assets, and marketValue columns
df['profits'] = df['profits'].apply(convert_to_common_unit)
df['assets'] = df['assets'].apply(convert_to_common_unit)
df['marketValue'] = df['marketValue'].apply(convert_to_common_unit)
df['revenue'] = df['revenue'].apply(convert_to_common_unit)
```

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
top_companies = df.sort_values(by='revenue', ascending=False).head(10)

# Plot the top 10 companies by revenue
plt.figure(figsize=(10, 6))
plt.barh(top_companies['organizationName'], top_companies['revenue'], color='skyblue')
plt.xlabel('Revenue (USD)')
plt.ylabel('Company')
plt.title('Top 10 Companies by Revenue')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest revenue at the top
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

```python
top_companies = df.sort_values(by='profits', ascending=False).head(10)

# Plot the top 10 companies by revenue
plt.figure(figsize=(10, 6))
plt.barh(top_companies['organizationName'], top_companies['profits'], color='green')
plt.xlabel('Profits (USD)')
plt.ylabel('Company')
plt.title('Top 10 Companies by Profits')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest revenue at the top
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

```python
top_companies = df.sort_values(by='assets', ascending=False).head(10)

# Plot the top 10 companies by revenue
plt.figure(figsize=(10, 6))
plt.barh(top_companies['organizationName'], top_companies['assets'], color='blue')
plt.xlabel(' Assets (USD)')
plt.ylabel('Company')
plt.title('Top 10 Companies by Assets')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest revenue at the top
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

```python
top_companies = df.sort_values(by='marketValue', ascending=False).head(10)

# Plot the top 10 companies by revenue
plt.figure(figsize=(10, 6))
plt.barh(top_companies['organizationName'], top_companies['marketValue'], color='blue')
plt.xlabel(' MarketValue (USD)')
plt.ylabel('Company')
plt.title('Top 10 Companies by MarketValue')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest revenue at the top
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

```python
# Group by 'organizationName' and 'country', summing the revenue for each company in each country
top_companies_by_country = df.groupby(['organizationName', 'country'])['revenue'].sum().reset_index()

# Sort the DataFrame by revenue in descending order
top_companies_by_country = top_companies_by_country.sort_values(by='revenue', ascending=False)

# Select the top 10 largest companies by revenue and country
top_10_by_country = top_companies_by_country.head(10)

# Plot the top 10 largest companies by revenue and country
plt.figure(figsize=(10, 6))
plt.barh(top_10_by_country['organizationName'] + ' - ' + top_10_by_country['country'],
         top_10_by_country['revenue'], color='skyblue')
plt.xlabel('Revenue (USD)')
plt.ylabel('Company - Country')
plt.title('Top 10 Largest Companies by Revenue and Country')
plt.gca().invert_yaxis()  # Invert y-axis to show the highest revenue at the top
plt.tight_layout()  # Adjust layout to prevent overlapping labels
plt.show()
```

```python
import pandas as pd
import dash
from dash import dcc, html, Input, Output
import plotly.express as px

app = dash.Dash(__name__)

company_options = [{'label': company, 'value': company} for company in df['organizationName'].unique()]

# Define the layout of the dashboard
app.layout = html.Div(children=[
    html.H1(children='Company Data Dashboard'),

    # Dropdown for selecting company name
    html.Label('Select Company Name:'),
    dcc.Dropdown(
        id='company-dropdown',
        options=company_options,
        value=company_options[0]['value']
    ),

    # Graph for displaying revenue, assets, profits, and market value
    dcc.Graph(id='company-metrics-graph')
])

# Define callback to update the graph based on selected company name
@app.callback(
    Output('company-metrics-graph', 'figure'),
    [Input('company-dropdown', 'value')]
)
def update_graph(selected_company):
    filtered_data = df[df['organizationName'] == selected_company]
    fig = px.bar(filtered_data, x='organizationName', y=['revenue', 'assets', 'profits', 'marketValue'],
                 barmode='group', title='Company Metrics')
    return fig

# Run the Dash app
if __name__ == '__main__':
    app.run_server(debug=True)
```
